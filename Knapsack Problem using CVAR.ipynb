```python
import gurobipy as gp
from gurobipy import GRB, quicksum
import numpy as np
import time
```

```python
a=[[6.1,5.93,8.66,10.28,5.13,6.33,8.6,10.1,7.53,8.63],
  [15.92,10.36,7.44,10.76,16.13,9.69,17.87,11.17,13.24,8.93],
  [15.99,12.71,12.27,16.24,8.07,19.74,10.8,9.34,10.82,13.17],
  [16.19,22.78,24.22,19.90,12.32,19.09,12.18,16.57,19.87,15.96],
  [19.21,24.14,19.42,28.63,27.66,28.45,25.38,23.67,25.11,29.59],
  [21.35,30.72,16.91,19.68,28.94,26.82,26.03,21.50,22.15,26.74]]
S=10
```

```python
#CVAR
tic=time.perf_counter()
knap=gp.Model()
x=knap.addVars(6,vtype=GRB.BINARY, name='x')
t=knap.addVar(vtype=GRB.CONTINUOUS, name='t')
Us=knap.addVars(10,vtype=GRB.CONTINUOUS, name='Us')
alpha=0.9
prob=0.1
knap.setObjective(3*x[0] +5*x[1]+6*x[2] +8*x[3]+10*x[4]+11*x[5],GRB.MAXIMIZE)

knap.addConstr(t+alpha**(-1)*quicksum(0.1*Us[s] for s in range(S))<=60)
for s in range(10):
    obj=0
    for i in range(6):
        obj+=a[i][s]*x[i] 

    knap.addConstr(Us[s]>=obj-t)
    knap.addConstr(Us[s]>=0)
    
    
knap.optimize()
xs=[]
for i in range(6):
    xs.append(x[i].x)

x=xs
knap.printAttr('x')
knap.objVal

toc=time.perf_counter()
print('Objective Value' , knap.objVal)
print('Runtime ' ,toc-tic)
print('Factories to Open :',x)
```

```python
#CVAR
S=6
tic=time.perf_counter()
knap=gp.Model()
x=knap.addVars(6,vtype=GRB.CONTINUOUS, name='x')
t=knap.addVar(vtype=GRB.CONTINUOUS, name='t')
Us=knap.addVars(6,vtype=GRB.CONTINUOUS, name='Us')
zminus=knap.addVars(6,vtype=GRB.INTEGER, name='zminus')
zplus=knap.addVars(6,vtype=GRB.INTEGER, name='zplus')
alpha=0.9
prob=0.1
knap.setObjective(3*x[0] +5*x[1]+6*x[2] +8*x[3]+10*x[4]+11*x[5] +10*prob*quicksum(zminus)+2*prob*quicksum(zplus),GRB.MINIMIZE)

knap.addConstr(t+alpha**(-1)*quicksum(0.1*Us[s] for s in range(S))==0)
for s in range(S):
    obj=0
    for i in range(6):
        obj+=x[i] 

    knap.addConstr(Us[s]>=obj+zplus[s]-zminus[s]-t)
    knap.addConstr(Us[s]>=0)
    
    
knap.optimize()
xs=[]
for i in range(6):
    xs.append(x[i].x)

x=xs
knap.printAttr('x')
knap.objVal

toc=time.perf_counter()
print('Objective Value' , knap.objVal)
print('Runtime ' ,toc-tic)
print('Factories to Open :',x)
```
