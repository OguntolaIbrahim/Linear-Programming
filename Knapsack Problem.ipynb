```python
import gurobipy as gp
from gurobipy import GRB, quicksum
import numpy as np
import time

```

```python
a=[[6.1,5.93,8.66,10.28,5.13,6.33,8.6,10.1,7.53,8.63],
  [15.92,10.36,7.44,10.76,16.13,9.69,17.87,11.17,13.24,8.93],
  [15.99,12.71,12.27,16.24,8.07,19.74,10.8,9.34,10.82,13.17],
  [16.19,22.78,24.22,19.90,12.32,19.09,12.18,16.57,19.87,15.96],
  [19.21,24.14,19.42,28.63,27.66,28.45,25.38,23.67,25.11,29.59],
  [21.35,30.72,16.91,19.68,28.94,26.82,26.03,21.50,22.15,26.74]]
```

```python

#EXPECTED VALUE SCENARIO
tic=time.perf_counter()
knap=gp.Model()
x=knap.addVars(6,vtype=GRB.BINARY, name='x')
knap.setObjective(3*x[0] +5*x[1]+6*x[2] +8*x[3]+10*x[4]+11*x[5],GRB.MAXIMIZE)


expected=np.sum(a,1)/10
obj=0
for i in range(6):
    obj+=expected[i]*x[i]
knap.addConstr(obj<=60)
knap.optimize()
expected
knap.printAttr('x')
xs=[]
for i in range(6):
    xs.append(x[i].x)

x=xs
toc=time.perf_counter()
print('Objective value :' ,knap.objVal)
print('Factories to Open : ',xs)
print('runtime :', toc-tic)
```

```python
print('SOLVING FOR MEAN SCENARIO')
print('Xs:',x)
print('Objective value',knap.objVal)
```

```python
statuses={}
for n in range(10):
    knap=gp.Model()

    knap.setObjective(3*x[0] +5*x[1]+6*x[2] +8*x[3]+10*x[4]+11*x[5],GRB.MAXIMIZE)


    obj=0
    for i in range(6):
        obj+=a[i][n]*x[i]

    knap.addConstr(obj<=60)
    knap.optimize()
    knap.printAttr('x')
    
    if knap.status==2:
        statuses[n]='feasible'
    if knap.status==3:
        statuses[n]='infeasible'
```

```python
#FOR EVERY SCENARIO
tic=time.perf_counter()
knap=gp.Model()
x=knap.addVars(6,vtype=GRB.BINARY, name='x')
knap.setObjective(3*x[0] +5*x[1]+6*x[2] +8*x[3]+10*x[4]+11*x[5],GRB.MAXIMIZE)

for s in range(10):
    obj=0
    for i in range(6):
        obj+=a[i][s]*x[i]

    knap.addConstr(obj<=60)
   
    
    
knap.optimize()

knap.printAttr('x')
knap.objVal
xs=[]
for i in range(6):
    xs.append(x[i].x)

x=xs
toc=time.perf_counter()
runtime=toc-tic
```

```python
print('SOLVING THAT SATISFIES ALL SCENARIOS')
print('Xs:',x)
print('Objective value',knap.objVal)
print('runtime : ',runtime)
```
